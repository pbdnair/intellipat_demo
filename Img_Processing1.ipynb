# THE MNIST DATABASE of handwritten digits
The MNIST database of handwritten digits, available from this page, has a training set of 60,000 examples, and a test set of 10,000 examples

<a href = 'http://yann.lecun.com/exdb/mnist/'> http://yann.lecun.com/exdb/mnist/ </a>




```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
(x_train, y_train),(x_test, y_test) = tf.keras.datasets.mnist.load_data() 
```

```python
compare = np.where(y_train==5) #this returns indexes of y_train wherever y ==5 in y_train
compare  # it has all indexes wherevr y_train=5
```

```python
type(compare)  #so this is list inside tuple ([1,2,3])
```

---  ---



```python
num = 10
print('X-Images from MNIST Dataset vs Y-their Labels (12 random samples per digit)')
for i in range(num):
  compare = np.where(y_train==i)  #pick tuple from y train for each values of i=0,1,2,.... one by one
  x = np.random.choice(compare[0],12)  #pick any 12 indexes wherever y_train = i = 0,1,2...
  count = 0
  plt.figure(figsize=(10,20))
  for j in x:
    count = count + 1
    #print(y_train[j])  This will print 000000000..111...222....9999999999
    plt.subplot(1,12,count)  
    plt.imshow(x_train[j,:,])   
    plt.xlabel(y_train[j])
    

```
